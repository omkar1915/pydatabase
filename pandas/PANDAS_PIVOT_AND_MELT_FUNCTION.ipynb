```python
from pandas import DataFrame
df = DataFrame(
              [
                ['Aryan','Singh','Full-time','administration',2,2000],
                ['Rohan','Agarval','intern','technical',3,5000],
                ['Riya','Shah','Full-time','administration',5,10000],
                ['Yash','Bhatia','Part-time','technical',7,10000],
                ['Siddant','Khanna','Full-time','Management',6,20000],
              ]
            , columns=['FIRST NAME','LAST NAME','TYPE','DEPARTMENT','YOE','SALARY']
              )
df
```

```python
import numpy as np
import pandas as pd
```

```python
table =pd.pivot_table(df,index ='TYPE')
table
```

```python
table =pd.pivot_table(df,index ='TYPE',columns='DEPARTMENT',values='SALARY',aggfunc=['sum','mean','count'])


table
```

```python
table =pd.pivot_table(df,index ='TYPE',columns='DEPARTMENT',values='SALARY',aggfunc='sum')


table
```
